```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("C:/Users/PRAJYOT UPADHYE/OneDrive/Desktop/ML/DATASETS/Social_Network_Ads.csv")
```

```python
df = df.iloc[:,2:]
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)
```

```python
X_train.shape , X_test.shape
```

# Standarscaler


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

# why scaling is important ?

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("actual Score:",accuracy_score(y_test,y_pred))
print("Scaled score:",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred=dt.predict(X_test)
y_pred_scaled=dt_scaled.predict(X_test_scaled)
```

```python
print("Acutual score:",accuracy_score(y_test,y_pred))
print("Scaled Score:",accuracy_score(y_test,y_pred_scaled))
```
